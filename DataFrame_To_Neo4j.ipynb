# Connecting to Neo4j 

```python
from py2neo import Graph, Node, Relationship
graph = Graph("bolt://localhost:11002",\
auth=("neo4j", "sayf"))
```

# Save data from Our dataframe to Neo4j database

```python
for row in merge3.iterrows():
    node_a = Node(str(row[1]["numImmatriculation"]), name=str(row[1]["numImmatriculation"]))
    node_d = Node(str(row[1]["numeroImmatriculationVehiculeAdverse"]), name=str(row[1]["numeroImmatriculationVehiculeAdverse"]))
    node_f = Node(str(row[1]["CompagnieAgence"]), name=str(row[1]["CompagnieAgence"]))
    node_b = Node(str(row[1]["sinistre_id"]), name=str(row[1]["numeroDuSinistre"]))
    node_e = Node(str(row[1]["codePostal"]), name=str(row[1]["new"]))


    rel = Relationship(node_a, "INVOLVED_IN", node_b)
    graph.merge(rel, "INVOLVED_IN", "name")
    rel2 = Relationship(node_d, "INVOLVED_IN", node_b)
    graph.merge(rel2, "INVOLVED_IN", "name")
    rel3 = Relationship(node_a, "LIVES_IN", node_e)
    graph.merge(rel3, "LIVES_IN", "name")
    rel4 = Relationship(node_a, "INSURED_AT", node_f)
    graph.merge(rel4, "INSURED_AT", "name")



```

```python
for row in merge3.iterrows():
    node_a = Node(str(row[1]["cga_vehicule_id"]), name=str(row[1]["numImmatriculation"]))
    node_b = Node(str(row[1]["sinistre_id"]), name=str(row[1]["numeroDuSinistre"]))
    rel = Relationship(node_a, "INVOLVED_IN", node_b)
    graph.merge(rel, "INVOLVED_IN", "name")
```
